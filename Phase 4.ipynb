```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

try:
    img = mpimg.imread("numpy_logo.png")

    plt.figure(figsize=(8, 4))

    # Original Logo
    plt.subplot(121)
    plt.imshow(img)
    plt.title("NumPy Logo")
    plt.axis("off")

    # Create Dark Logo
    dark_logo = img.copy()

    if img.shape[2] == 4:  # RGBA image
        dark_logo[:, :, :3] = 1 - dark_logo[:, :, :3]
    else:  # RGB image
        dark_logo = 1 - dark_logo

    # Display Dark Logo
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("NumPy Dark Logo")
    plt.axis("off")

    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print("NumPy logo not found")

except Exception as e:
    print("Error:", e)
```
