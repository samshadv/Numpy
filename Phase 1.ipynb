### Creating Array fron scratch 

```python
zeros  = np.zeros((2,5))
print("Zeros Array: ",zeros)

ones  = np.ones((3,5))
print("ones Array: ",ones)

full = np.full((2,2),8)
print("full Array: ",full)

random = np.random.random((2,3))
print("random Array: ",random)

sequence = np.arange(0,11,2)
print("Sequence Array: ",sequence)
```

### Vector , matrix, tensor


```python
vector = np.array([1,2,3])
print("v",vector)

matrix = np.array([[1,2,3],[4,5,6]])
print("m",matrix)

tensor = np.array([[[1,2],[3,4]],
                   [[1,2],[4,5]]])
print("t",tensor)
                    
```

### Array Reshapping

```python
arr = np.arange(20)
print("Original Array:", arr)

reshaped = arr.reshape((4,5))
print("Reshaped Array:\n", reshaped)

flattened = reshaped.flatten() # return copy
print("Flattened Array:", flattened)

raveled = reshaped.ravel() # return view , insted of copt
print("Raveled Array:", raveled)
```

### Transpose


```python
transpose = reshaped.T
print(transpose)
```
