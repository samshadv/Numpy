### np.concatenate

numpy.concatenate() function concatenate a sequence of arrays along an existing axis.

```python
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)

print(c)
print(d)
```

```python
np.concatenate((c,d))
```

```python
np.concatenate((c,d),axis=1)
```

### Set functions

- np.union1d
- np.intersect1d
- np.setdiff1d
- np.setxor1d
- np.in1d

```python
m = np.array([1,2,3,4,5])
n = np.array([3,4,5,6,7])

np.union1d(m,n)
```

```python
np.intersect1d(m,n)
```

```python
np.setdiff1d(m,n)
# whic are in a but not in b
```

```python
np.setxor1d(m,n)
```

```python
m[np.in1d(m,1)]
```
