```python
import numpy as np

sales_data = np.array([
    [1,  150000, 180000, 220000, 250000],   # Paradise Biryani
    [2,  120000, 140000, 160000, 190000],   # Beijing Bites
    [3,  200000, 230000, 260000, 300000],   # Pizza Hub
    [4,  180000, 210000, 240000, 270000],   # Burger Point
    [5,  160000, 185000, 205000, 230000],   # Chai Point
    [6,   95000, 125000, 155000, 185000],   # Vada Pav King
    [7,  135000, 165000, 195000, 235000],   # Misal Pavilion
    [8,  210000, 245000, 280000, 320000],   # Dosa Palace
    [9,  110000, 130000, 170000, 210000],   # Kathi Junction
    [10, 175000, 200000, 235000, 265000]   # Tandoor Nights
    
])

print("Shape of sample data:", sales_data.shape)
print("total of data", sales_data[:,1:])
```

```python
# total sales per year
print("Sales per year\n",np.sum(sales_data, axis = 0))
yearly_total = np.sum(sales_data[:,1: ], axis = 0)
print("\n",yearly_total)

# Min sales per restro
min_resto = np.min(sales_data[:,1:],axis = 1)
print("minimun per resto \n",min_resto)

#max Sales per year

max_sales =np.max(sales_data[:,1:], axis = 0)
print("max sales per year:\n", max_sales)

# Average Annual Sales

average = np.mean(sales_data[:,1:], axis =1)
print("Average sales\n",average)


```

## axis=0 → Operates down the rows (column-wise)
## axis=1 → Operates across the columns (row-wise)

```python
# Cumulative Sales Per Restaurant (Running Total)
cumsum = np.cumsum(sales_data[:,1:], axis =1)
print(cumsum)
```

```python
plt.figure(figsize=(8, 6))

# Plot average cumulative sales across all restaurants
plt.plot(np.mean(cumsum, axis=0), linewidth=2)

plt.title("Average Cumulative Sales Across All Restaurants", fontsize=14)
plt.xlabel("Years", fontsize=12)
plt.ylabel("Sales", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector addition", vector1 + vector2)

print("vector multiplication", vector1 * vector2)

print("\n Dot product", np.dot(vector1,vector2))

angle = np.arccos(np.dot(vector1,vector2)/ (np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print(angle)
```

```python
#vectorized operation 
restorent = np.array(['biryani', 'chines', 'pizza', 'burger'])
vec_upper = np.vectorize(str.upper)
vect_title = np.vectorize(str.title)

print(vec_upper(restorent))
print(vect_title(restorent))

def add_emoji(food):
    return food + " 🍽️"

emoji_func = np.vectorize(add_emoji)
print(emoji_func(restorent))
```
